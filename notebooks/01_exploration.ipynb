```python
engine.dispose()
```

```python
query = """
SELECT 
    o.order_id,
    o.customer_id,
    o.order_purchase_timestamp,
    SUM(p.payment_value) AS revenue
FROM orders o
JOIN payments p
ON o.order_id = p.order_id
GROUP BY 
    o.order_id,
    o.customer_id,
    o.order_purchase_timestamp;
"""
```

```python
df_fact = pd.read_sql(query, engine)
```
